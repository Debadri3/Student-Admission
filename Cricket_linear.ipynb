##Data Exploration (Part II)
Here I use get dummies function to convert objects to categorical features

```
one_hot_data = pd.concat([df1, pd.get_dummies(df1['Bats(handed)'], prefix='Batting hand')], axis=1)
one_hot_data = one_hot_data.drop('Bats(handed)', axis=1)
one_hot_data
```

```
one_hot_data2 = pd.concat([one_hot_data, pd.get_dummies(df1['Bowls(handed)'], prefix='Bowling hand')], axis=1)
one_hot_data2 = one_hot_data2.drop('Bowls(handed)', axis=1)
one_hot_data2
```

```
one_hot_data2["Age"].max()
```

```
#Now I'll use MinMaxScaler on the data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(one_hot_data2)
new_data=scaler.transform(one_hot_data2)
processed_data=pd.DataFrame(new_data,columns=one_hot_data2.columns)
processed_data

```

```
processed_data["Age"]
```

```
processed_data["Age"]=(processed_data["Age"]*85)+1
```

```
processed_data
```

```
from sklearn.model_selection import train_test_split
X=processed_data.drop("Age",axis=1)
y=processed_data["Age"]
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```
X_test.shape
```

```
import xgboost as xgb
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso,Lars,SGDRegressor
from sklearn.svm import SVR,LinearSVR
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np
# Put models in a dictionary
models = {"Lasso": Lasso(),
          "LARS": Lars(), 
          "SGD": SGDRegressor(), "XGB": xgb.XGBRegressor(), "K Neighbours":KNeighborsRegressor(), "Gradient Boosting":GradientBoostingRegressor(),"SVC":SVR(), "Linear SVC":LinearSVR()}



# Create function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):

    
    np.random.seed(42)
    
    model_scores = {}
    
    for name, model in models.items():
        
        model.fit(X_train, y_train)
        
        model_scores[name] = model.score(X_test, y_test)*100
    return model_scores
```

```
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)
model_scores
```
